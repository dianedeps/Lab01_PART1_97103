### 02. Raw to Silver ingestion

Read all files in raw folder and transform them to parquet files in the silver layer

```python
from pathlib import Path
import sys

project_root = Path.cwd().parent
sys.path.append(str(project_root))
```

```python
from src.ingestion.raw_to_silver_ingestion import RawToSilverIngestion
```

```python
raw_dir = project_root / "data" / "raw"
silver_dir = project_root / "data" / "silver"
metadata_dir = silver_dir / "metadata"

print(f"Raw dir: {raw_dir}")
print(f"Silver dir: {silver_dir}")
print(f"Metadata dir: {metadata_dir}")
```

```python
ingestion = RawToSilverIngestion(
    raw_dir=str(raw_dir),
    silver_dir=str(silver_dir),
    metadata_dir=str(metadata_dir),
)

ingestion.run()
```

```python
print("Silver files:")
for file_path in sorted(silver_dir.glob("*.parquet")):
    print(f" - {file_path.name}")

print("\nMetadata files:")
for file_path in sorted(metadata_dir.glob("*.json")):
    print(f" - {file_path.name}")
```
